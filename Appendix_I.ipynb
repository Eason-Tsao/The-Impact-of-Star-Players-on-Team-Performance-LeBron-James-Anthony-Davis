```python
# Combine all the data into one dataframe
import pandas as pd
df_combined = pd.concat([year2017_2018, year2018_2019, year2019_2020], ignore_index=True)
print(f"Combined DataFrame has {df_combined.shape[0]} rows.")
df_combined['date'] = pd.to_datetime(df_combined['date'])
```

```python
# To find power partnership in away comp from 2019-2020
import pandas as pd
year2019_2020['score_diff'] = year2019_2020['away_score'] - year2019_2020['home_score']
away_leads = year2019_2020[year2019_2020['score_diff'] > 0]
away_players_columns = ['a1', 'a2', 'a3', 'a4', 'a5']
away_leads_selected = away_leads[away_players_columns + ['score_diff']]
player_positive_diff_counts = pd.Series(dtype=int)
for player_col in away_players_columns:
    player_counts = away_leads_selected[player_col].value_counts()
    player_positive_diff_counts = player_positive_diff_counts.add(player_counts, fill_value=0)
player_positive_diff_counts_sorted = player_positive_diff_counts.sort_values(ascending=False)
player_positive_diff_counts_sorted
```

```python
# To find power partnership in home comp rom 2019-2020
import pandas as pd
year2019_2020['score_diff1'] = year2019_2020['home_score'] - year2019_2020['away_score']
home_leads = year2019_2020[year2019_2020['score_diff1'] > 0]
home_players_columns = ['h1', 'h2', 'h3', 'h4', 'h5']
home_leads_selected = home_leads[home_players_columns + ['score_diff1']]
player_home_positive_diff_counts = pd.Series(dtype=int)
for player_col in home_players_columns:
    player_counts = home_leads[player_col].value_counts()
    player_home_positive_diff_counts = player_home_positive_diff_counts.add(player_counts, fill_value=0)
player_home_positive_diff_counts_sorted = player_home_positive_diff_counts.sort_values(ascending=False)
player_home_positive_diff_counts_sorted
```

```python
# Points earned by LeBron James and Anthony Davis
import pandas as pd
import matplotlib.pyplot as plt
# Ensure matplotlib shows our plots
%matplotlib inline
# Define the players to analyze
players_to_analyze = ['LeBron James', 'Anthony Davis']
# Create a total points column
df_combined['total_points'] = df_combined['home_score'] + df_combined['away_score']
# Check if either player is in any of the starting position columns for either team
df_filtered = df_combined[
    df_combined['a1'].isin(players_to_analyze) |
    df_combined['a2'].isin(players_to_analyze) |
    df_combined['a3'].isin(players_to_analyze) |
    df_combined['a4'].isin(players_to_analyze) |
    df_combined['a5'].isin(players_to_analyze) |
    df_combined['h1'].isin(players_to_analyze) |
    df_combined['h2'].isin(players_to_analyze) |
    df_combined['h3'].isin(players_to_analyze) |
    df_combined['h4'].isin(players_to_analyze) |
    df_combined['h5'].isin(players_to_analyze)]
# Pivot table to summarize the total points by game
pivot_points = pd.pivot_table(
    df_filtered,
    values='total_points',
    index='game_id',
    aggfunc='mean')
# Plot the results
pivot_points.plot(kind='bar')
plt.show()
```
